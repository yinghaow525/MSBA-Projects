## Use "Beautiful Soup" Library to implement an "html/xml parser"

Here is the link to the documentation:
https://www.crummy.com/software/BeautifulSoup/bs4/doc/

In this example we will download the Income Statement for Apple Inc. (AAPL) from the sec website.


```
from bs4 import BeautifulSoup
soup = BeautifulSoup(page.content, 'lxml')
```

```
# The Prettify function formats that html/xml text to make it easier to read.
# You would inspect the html code to find marker tags that identifies the specific section of text you wish to extract.
print(soup.prettify())
```

## The Next Step is to Find the HTML Tags


### The tag  ***\<td class="nump"*** precedes any reported numbers AAPL's financial statements 

```
# using re to match the exact format of desired item & numbers
import re

# create rules to extract only numbers and item
# only one value for each match, no need to use greedy search (?)
findnum = re.compile(r'<td class="nump">(.*)<span>')
finditem = re.compile(r'>(.*)</a>')
# notice: special rows
# eg: <strong>Shares used in computing earnings per share:</strong>
```

```
# Find the tag that identify numbers from Apple's Income Statement 
Numbers = soup.find_all('td', class_='nump')
```

### Print out the scraped Numbers that had the class tag "nump" from  Apple's Income Statement 

```
Numbers
```

### Use re to find the specified string stored in Numbers object - we want only numbers

```
# use findall() & pre-set rule to find the specified string through regular expressions
number = []
for num in Numbers:
  # [0]: used to return a single element, otherwise the object will be a list
  num = re.findall(findnum, str(num))[0]
  # append to number list
  number.append(num)
```

```
# number length --> total 19 * 3 non null values
len(number)
```

```
# first 6 element to ensure the format
number[:6]
```

### The tag ***\<a class="a"*** precedes reported items of AAPL's Income Statement

```
# find the tag that identify items from Income Statement from Apple's Income Statement
Items = soup.find_all('a', class_= 'a')
```

### Print out the scrapped Items that had the class tag "a" from  Apple's Income Statement 

```
# Find the tag that identify numbers from Apple's Income Statement 
Items
```

### Use re to find the specified string stored in Items object

```
item = []
for i in Items:
  # [0]: used to return a single element, otherwise the object will be a list
  i = re.findall(finditem, str(i))[0]
  # append to item list
  item.append(i)
```

```
# item length --> 5 more rows than number non-null values
len(item)
```

```
# print out result
# some items have uncleared tags, but will be used for further df construction
item
```

## Construct Income Statement DataFrame

```
# import libraries
import pandas as pd
import numpy as np
pd.options.mode.chained_assignment = None
```

### Split numbers into 3 columns

one column for each year.

notice the sequence of numbers is looping: 2021, 2020, 2019, 2021, 2020, 2019 ... , thus we can use a for loop to write into 3 lists with specified index.
- if index%3 == 0: number value is for 2021
- if index%3 == 1: number value is for 2020
- if index%3 == 2: number value is for 2019

```
# split numbers into 3 columns, one column for each year
# notice the sequence of numbers is looping: 2021, 2020, 2019 ... 
Num_2021 = []
Num_2020 = []
Num_2019 = []

for i in range(len(number)):
  if i%3 == 0:
    Num_2021.append(number[i])
  elif i%3 == 1:
    Num_2020.append(number[i])
  else:
    Num_2019.append(number[i])
```

### DataFrame initialization with (24, 4) shape
- first column: Income Statement Items for Apple
- other 3 columns: temporarily filled with NaNs

```
# construct the Dataframe
# cannot directly use Num_xxxx list because length does not match
AAPL_IncomeStatement = pd.DataFrame({'Income Statement Items for Apple': item,
                                    #  'Sept 25 2021': Num_2021, 
                                    #  'Sept 26 2020': Num_2020,
                                    #  'Sept 28 2019': Num_2019
                                     })
# instead, add in new columns and write in NaN values
AAPL_IncomeStatement['Sept 25 2021'] = np.nan
AAPL_IncomeStatement['Sept 26 2020'] = np.nan
AAPL_IncomeStatement['Sept 28 2019'] = np.nan
```

```
# brief look at current DataFrame, make sure the shape is (24, 4) 
AAPL_IncomeStatement.shape
```

### Write in numbers for all 3 years, depending on item value pattern

```
j = 0
# find special pattern of item values (which do not have corresponding number value in that row)
matches = [':','Product','Service']
for i in range(len(AAPL_IncomeStatement)):
  # find rows where numbers should be none
  if any(x in AAPL_IncomeStatement['Income Statement Items for Apple'][i] for x in matches):
    AAPL_IncomeStatement['Sept 25 2021'][i] = np.nan
    AAPL_IncomeStatement['Sept 26 2020'][i] = np.nan
    AAPL_IncomeStatement['Sept 28 2019'][i] = np.nan
  # write in numbers for non-NaN rows
  else:
    AAPL_IncomeStatement['Sept 25 2021'][i] = Num_2021[j]
    AAPL_IncomeStatement['Sept 26 2020'][i] = Num_2020[j]
    AAPL_IncomeStatement['Sept 28 2019'][i] = Num_2019[j]
    # j increments only when numbers are successfully written in, 
    j += 1
```

### Remove remained HTML tags in Item column value

```
# remove <strong>, </strong> for item value
AAPL_IncomeStatement['Income Statement Items for Apple'] = AAPL_IncomeStatement['Income Statement Items for Apple'].apply(lambda x: 
                                                                                                                          x.replace('<strong>',''))
AAPL_IncomeStatement['Income Statement Items for Apple'] = AAPL_IncomeStatement['Income Statement Items for Apple'].apply(lambda x: 
                                                                                                                          x.replace('</strong>',''))
```
